# Load AES Data

```python
# Define Design Characteristics
aes_clk_period    = 10
aes_ylim          = 500
aes_num_trials    = 25
root_aes_data_dir = '/Users/ti27457/Repos/ttb/circuits/aes/tjfree_75eds_seed'

# List of data frames
aes_dfs = []

# Load dataframes
for i in range(aes_num_trials):
    print "Loading seed %d data..." % (i)
    aes_dfs.append(load_data_df_wf(root_aes_data_dir + str(i), aes_clk_period, 'aes_seed' + str(i)))
print "Done."

# Concatenate dataframes
merged_aes_df = pd.concat(aes_dfs)
```

```python
# Plot Settings
LINE_WIDTH = 2

# Create Figure
sns.set()
fig, ax = plt.subplots(1, 1, figsize=(9, 3))

# Plot Data
sns.lineplot(x="Time", y="Total Malicious Coalesced TTTs",   data=merged_aes_df, ax=ax, linewidth=LINE_WIDTH, estimator=None)
sns.lineplot(x="Time", y="Total Malicious Distributed TTTs", data=merged_aes_df, ax=ax, linewidth=LINE_WIDTH, estimator=None)

# Format Main Plot
ax.set_ylim(-10, aes_ylim)
ax.set_xlabel('Clock Cycles')
ax.set_ylabel('# SSCs')
for tick in ax.get_yticklabels():
    tick.set_rotation(90)

# Save Plot as PNG
plt.savefig("aes_fuzzing_1.pdf", format='pdf', bbox_inches='tight', transparent=False)
```
